# Storing DNA Sequences

DNA sequences can be stored as strings in Python. A string is a sequence of characters enclosed within quotes (' ' or " ").

<br><br><br>

> Let's take a simple DNA sequence, "ATGCTAGC", and store it in a variable.


<br>

```python
# Storing a DNA sequence
dna_seq = "ATGCTAGC"

# Print the sequence
print(dna_seq)

```
-----

<br><br><br>

You can access specific parts of the sequence and manipulate them as needed.

<br>

# Analyzing a DNA Sequence

Let's start with a DNA sequence and perform the following tasks:

* Count the occurrences of each nucleotide (A, T, G, C).
* Calculate the length of the DNA sequence.
* Calculate the GC content (percentage of G and C nucleotides in the sequence).



```python
# Storing a DNA sequence
dna_sequence = "ATGCTAGCGCGTAA"

# Counting occurrences of each nucleotide
count_A = dna_sequence.count('A')
count_T = dna_sequence.count('T')
count_G = dna_sequence.count('G')
count_C = dna_sequence.count('C')

# Printing counts
print("A:", count_A)
print("T:", count_T)
print("G:", count_G)
print("C:", count_C)

# Calculating the length of the DNA sequence
sequence_length = len(dna_sequence)
print("Length of DNA Sequence:", sequence_length)

# Calculating GC content
gc_content = ((count_G + count_C) / sequence_length) * 100
print("GC Content (%):", gc_content)

```
